```python
import datetime
import webbrowser
import subprocess
from reportlab.pdfgen import canvas
from reportlab.pdfgen.canvas import *
from reportlab.pdfbase import pdfmetrics
from reportlab.pdfbase.ttfonts import TTFont


import pytz
import openpyxl
from openpyxl import *
import xlwt 
from xlwt import Workbook 
precios= {"Hamburguesa Sencilla":3000,"Hamburguesa Especial":7000,"Papa Frita":3000,"Yuca Frita":3000,
                       "Queso":3000,"Salchipapa Sencilla":3000,"Salchipapa Especial":6000,"Salvajada":20000,"Perro Sencillo":3000,
                       "Perro Especial":6000,"Coca Cola":4000,"Pepsi":3000,"Aguila":4500}
        
productos={"Hamburguesa Sencilla":1,"Hamburguesa Especial":1,"Papa Frita":1,"Yuca Frita":1,
                       "Queso":1,"Salchipapa Sencilla":1,"Salchipapa Especial":1,"Salvajada":1,"Perro Sencillo":1,
                       "Perro Especial":1,"Coca Cola":1,"Pepsi":1,"Aguila":1}

def serial(numero):
    s= "000000"
    resultado= s[0:len(s)-len(str(numero))] + str(numero)
    return resultado
```

```python
#Creating a pdf file and setting a naming convention
def hacer_pedido(nombre, productos, precios):
    pedido=1
    
    productos_selec={}
    for i in productos:
        if productos[i]>0:
            productos_selec[i]=productos[i]
    y=30
    pdfmetrics.registerFont(TTFont('Arial', 'Arial.ttf'))
    c = canvas.Canvas("invoice.pdf",pagesize=(300,230+(len(productos_selec)*40)),bottomup=0)
    c.setFont('Arial',15)
    text = "Recibo Electrónico"
    text_width = c.stringWidth(text,'Arial',15)
    c.drawString((300-text_width)/2, y, text)
    print((300-text_width)/2)
    #Separador
    y=y+20
    text="------------------------------------------------------------"
    c.drawString(0, y, text)
    #Info Negocio
    c.setFont('Arial',9)
    y=y+20
    text="Dirección: Diagonal 21 No. 29-56, Barrio Sabanas del Valle"
    c.drawString(20, y, text)
    y=y+20
    text="Telefono: 312 5698741"
    c.drawString(20, y, text)
    #Separador
    c.setFont('Arial',15)
    y=y+20
    text="------------------------------------------------------------"
    c.drawString(0, y, text)
    #Separador
    c.setFont('Arial',9)
    y=y+20
    text="Fecha: "+ datetime.datetime.now(pytz.timezone('America/Bogota')).strftime("%m/%d/%Y, %H:%M:%S")
    c.drawString(20, y, text)
    c.setFont('Arial',9)
    y=y+20
    text="Nombre de Cliente: "+nombre
    c.drawString(20, y, text)
    y=y+20
    text="ID del Pedido: "+ serial(pedido)
    c.drawString(20, y, text)
    i=0
    c.setFont('Arial',8)
    for j in productos_selec:
        y=y+20
        text= "["+str(i+1)+"]"+" "+j+" x "+str(productos_selec[j])
        c.drawString(20, y, text)
        y=y+20
        text= "Precio: "+str(precios[j]*productos_selec[j]) + " $"
        c.drawString(20, y, text)
        i+=1
    #Separador
    c.setFont('Arial',15)
    y=y+20
    text="------------------------------------------------------------"
    c.drawString(0, y, text)
    #Total a pagar
    # Saving the PDF
    c.setFont('Arial',11)
    y=y+20
    text="Total A Pagar: "+str(sum([productos_selec[i]*precios[i] for i in productos_selec])) + " $"
    c.drawString(20, y, text)
    c.showPage()
    c.save()
    chrome_path= "C:\Program Files (x86)\Google\Chrome\Application\chrome.exe"
    subprocess.Popen([chrome_path, "invoice.pdf"])
    # Workbook is created 
    wb=load_workbook("registro.xlsx")
    ws=wb["Hoja de pedidos"] 
    #Valores
    ws.cell(pedido+1, 1).value=serial(pedido)
    ws.cell(pedido+1, 2).value= datetime.datetime.now(pytz.timezone('America/Bogota')).strftime("%m/%d/%Y, %H:%M:%S")
    ws.cell(pedido+1, 3).value=nombre
    contador=0
    for i in productos_selec:
        ws.cell(pedido+1,  2+contador+2).value=productos_selec[i]
        contador+=1
        
    ws.cell(pedido+1, 17).value=sum([productos_selec[i]*precios[i] for i in productos_selec])
    wb.save('registro.xlsx') 
    
    
hacer_pedido("Alberto Mendoza",productos,precios)
```
